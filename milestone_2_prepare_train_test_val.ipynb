```python
import numpy as np
import pandas as pd
```

```python
%%time
df_milestone2 = pd.read_pickle("data\\Milestone_II_data.pkl")
```

```python
df_milestone2.shape
```

```python
df_milestone2
```

```python
X_train = df_milestone2.loc[df_milestone2['Date'] < '2020-06-20']
X_train.shape
```

```python
df_test = df_milestone2.loc[(df_milestone2['Date'] > '2020-07-20') & (df_milestone2['Date'] < '2021-07-23')]
df_test.shape
```

```python
y_train = X_train['y_20d']
y_test_full = df_test['y_20d']
X_train = X_train.drop(['y_20d'], axis=1)
df_test_full = df_test.drop(['y_20d'], axis=1)
```

Split the Test data into a seen (test) and an unseen (validation) portion. This is as per Kaggle competition setups. Final validation on the unseen portion will only be done once at the very end.

```python
from sklearn.model_selection import train_test_split
```

```python
X_test, X_val, y_test, y_val = train_test_split(df_test_full, y_test_full, test_size=0.5, random_state=42)
```

# Save all the dataframes
We will save the train and test data to data files. We can retrieve them from here later rather than rerunning the pipeline. We will save the validation as well but only use only load at the end for final model validation.

```python
%%time
X_train.to_pickle("data\\milestone_data_X_train.pkl")
y_train.to_pickle("data\\milestone_data_y_train.pkl")

X_test.to_pickle("data\\milestone_data_X_test.pkl")
y_test.to_pickle("data\\milestone_data_y_test.pkl")

X_val.to_pickle("data\\milestone_data_X_val.pkl")
y_val.to_pickle("data\\milestone_data_y_val.pkl")
```
